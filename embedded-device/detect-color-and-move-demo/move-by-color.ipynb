```python
import cv2
import numpy as np
import traitlets
from IPython.display import display
import ipywidgets.widgets as widgets
from jetbot import Camera, bgr8_to_jpeg
from ColorDetector import ColorDetector
from Drawbot import Drawbot
import time
```

```python
drawbot = Drawbot()
camera = Camera.instance(width=224, height=224)
image = widgets.Image(format='jpeg', width=224, height=224)
color_detector = ColorDetector()
```

```python
camera_link = traitlets.dlink((camera, 'value'), (image, 'value'), transform=bgr8_to_jpeg)
display(widgets.VBox([widgets.HBox([image])]))

def execute(change):
    image = change['image']
#     color_name = color_detector.get_color_name(image)
    color_name = color_detector.detect_color(image)
    print(color_name)
    drawbot.step_forward()
    if color_name == "Red":
        drawbot.stop()
        drawbot.step_left()
        drawbot.wait(1.68)
        drawbot.stop()
    elif color_name == "Green":
        drawbot.stop()
        drawbot.step_left(0.4)
        drawbot.wait(5.68)
        drawbot.stop()
    elif color_name == "Blue":
        drawbot.stop()
        drawbot.step_right()
        drawbot.wait(1)
        drawbot.stop()
        
    print(color_name, end=' ')    
```

```python
while(True):
    time.sleep(1.0)
    execute({'image': camera.value})
```

```python
drawbot.stop()
```
